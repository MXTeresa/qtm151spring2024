(a) Functions

<font size = "3">

- Read the dataset labelled features.csv.
- Create a ``` lambda ``` function that checks if horsepower $>$ 100
- Create a new list that shows true if horsepower $>$ 100 and false <br>
if horsepower $\leq$ 100.
- Count how many cars have horsewpower $> 100$.

To get full credit use either the apply or the map function <br>
(see Lectures 7 and 8)

````python
# Write your answer here

features = pd.read_csv("Data/features.csv")
features.replace("?", np.nan, inplace = True) # Replace "?" with NaN
features["horsepower"] = pd.to_numeric(features["horsepower"]) # transform the horsepower column to numeric column

check_horsepower = lambda hp: hp > 100

horsepower_gt_100 = features["horsepower"].apply(check_horsepower)
count_gt_100 = horsepower_gt_100.sum()
print("Number of cars with horsepower > 100: " + str(count_gt_100))
````

(b) Subsetting Data

<font size = "3" >

- Read the dataset labelled features.csv.
- Define a global variable called ```threshold``` that takes the <br>
value 3000
- Subset the data to get one dataset with weight $>$ threshold and <br>
another with weight $\le$ 3000. Use the ```.query()``` function.
- Compute the mean acceleration for each of these datasets.<br>

Note: Review how to define a gloabl variable inside ```.query()```
 

````python
# Write your answer here:

threshold = 3000

features_gt_threshold = features.query("weight > @threshold")
features_le_threshold = features.query("weight <= @threshold")

mean_acceleration_gt_threshold = features_gt_threshold["acceleration"].mean()
mean_acceleration_le_threshold = features_le_threshold["acceleration"].mean()

print("Mean acceleration for weight > threshold: " + str(mean_acceleration_gt_threshold))
print("Mean acceleration for weight <= threshold: " + str(mean_acceleration_le_threshold))
````

(c) Regression

<font size = "3">

In this exercise you are asked to plot acceleration vs. weight and to plot <br>
 the best fit line.

- Run a linear regression between acceleration and weight.
- Extract the intercept and coefficient.
- Using your predicted coefficients, compute the predicted value of acceleration
- Plot a scatter plot with acceleration on the y axis and weight on the x axis.
- On that same figure, plot a line with the predicted values of acceleration

Do not forget to label your axes and title.

````python
# Write your answer here:

model = smf.ols(formula = "acceleration ~ weight", data = features) # run a linear regression
result = model.fit()

b_list = result.params # extract intercept and coefficient
print(b_list)

predicted_acceleration = b_list[0] + b_list[1] * features["weight"] # predicted values of acceleration

plt.scatter(x = features["weight"], y = features["acceleration"], label = "Data")
plt.plot(features["weight"], predicted_acceleration, color = "red", label = "Best Fit Line")
plt.xlabel("Weight")
plt.ylabel("Acceleration")
plt.title("Acceleration vs. Weight")
plt.legend()
plt.show()
````
